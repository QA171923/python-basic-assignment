5. Explain the concept of type casting in Python with examples?
- Implicit Casting: Done automatically by Python (e.g., converting int to float).
# Example:
 num1=10
 num2=1.5
 result=num1+num2
 print(result)
 print(type(result))

- Explicit Casting: Done manually by the programmer using type conversion functions like int(), float(), str(), etc.
# Example:
 num1=3.4
 num2=int(num1)
 print("original number=",num1)
 print("number after typecasting=",num2)
 print(type(num2))
